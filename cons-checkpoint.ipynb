```python
from sklearn.model_selection import train_test_split #
```

```python
x=  df_raw.drop("Strength",axis=1) # independent dataset
y=  df_raw["Strength"] # dependent dataset
```

```python
#feature selection useing mutual info technique
from sklearn.feature_selection import f_regression, mutual_info_regression
a=mutual_info_regression(x,y,discrete_features='auto', n_neighbors=2, copy=True, random_state=None)
df_inportance =pd.DataFrame(a,columns=['inportance'])
c=pd.DataFrame(x.columns)
fs = pd.concat([c,df_inportance],axis=1)
fs
# flyash has less influence to dependent variable strength.
```

```python
 X_train, X_test, y_train, y_test = train_test_split(
  x.drop("Fly Ash",axis=1), y, test_size=0.30, random_state=0) #preare test-train dataset without considering flyash
```

```python
#prepare mode : polynomial regression
#"To overcome under-fitting, we need to increase the complexity of the model,
#This is still considered to be linear model as the coefficients/weights associated with the features are still linear. 
#x2,x3 is only a feature. However the curve that we are fitting is quadratic in nature.
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import explained_variance_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import max_error
from sklearn.pipeline import Pipeline
import numpy as np
#initilization of degree of polynimial feature,
d=[1,2,3]
for i in d:
 model = Pipeline([('poly', PolynomialFeatures(degree=i)),
               ('linear', LinearRegression(fit_intercept=False))]) #model prepparation

 
 m=model.fit( X_train,y_train) #fit on training dataset
 
 y_pred = model.predict(X_test) #apply the model on x_test and pradict Y_test
    
 #compparision between    
 score = r2_score(y_test,y_pred)
 score1 =max_error(y_test, y_pred)
 score2 =mean_squared_error(y_test, y_pred)
 
 print(score1,score2,score)
```

```python
#Model performance with respect to RSME and R2 score,

final_model = Pipeline([('poly', PolynomialFeatures(degree=3)),
               ('linear', LinearRegression(fit_intercept=False))])
    
final_model.fit( X_train,y_train)

# predicting on training data-set
y_train_predicted = final_model.predict(X_train)
  
#predicting on test data-set
y_test_predict =final_model.predict(X_test)
 
# evaluating the model on test dataset
rmse_train = np.sqrt(mean_squared_error(y_train, y_train_predicted))
r2_train = r2_score(y_train, y_train_predicted)
  
# evaluating the model on test dataset
rmse_test = np.sqrt(mean_squared_error(y_test, y_test_predict))
r2_test = r2_score(y_test, y_test_predict)

print("RMSE OF TRAIN SET",rmse_train, '&' ,"RSME OF TEST SET",rmse_test)
print("R2 OF TRAIN SET",r2_train,'&' ,"R2 OF TEST SET",r2_test)

print("% of error of rmse =",(rmse_test-rmse_train)/rmse_test)
print("% of error of r2 =",(r2_test-r2_train)/r2_test)
```
